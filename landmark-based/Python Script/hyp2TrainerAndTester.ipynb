```python
IMG_SIZE = 256        # input resolution
NUM_LANDMARKS = 4     # OFD1, OFD2, BPD1, BPD2
HEATMAP_SIZE = 256    
GAUSSIAN_RADIUS = 4
```

```python
import numpy as np
import cv2

def generate_heatmap(h, w, x, y, radius=4):
    heatmap = np.zeros((h, w), dtype=np.float32)
    
    x = int(x)
    y = int(y)

    if x < 0 or y < 0 or x >= w or y >= h:
        return heatmap  
    
    cv2.circle(heatmap, (x, y), radius, 1.0, -1)
    heatmap = cv2.GaussianBlur(heatmap, (0,0), sigmaX=radius, sigmaY=radius)
    heatmap = heatmap / heatmap.max() if heatmap.max() > 0 else heatmap
    return heatmap
```

```python
def resize_with_landmarks(img, landmarks):
    h, w = img.shape[:2]

    img_resized = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    scale_x = IMG_SIZE / w
    scale_y = IMG_SIZE / h

    scaled = []
    for x, y in landmarks:
        scaled.append((x * scale_x, y * scale_y))

    return img_resized, scaled
```

```python
train_augs = A.Compose([
    A.Rotate(limit=20, p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1,
                       rotate_limit=10, border_mode=cv2.BORDER_REFLECT_101, p=0.5),
], keypoint_params=A.KeypointParams(format='xy'))
```

```python
import torch
from torch.utils.data import Dataset

class FetalLandmarkDataset(Dataset):
    def __init__(self, df, img_dir, augment=False):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.augment = augment

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        try:
            row = self.df.iloc[idx]
            img_path = os.path.join(self.img_dir, row['image_name'])

            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                raise ValueError("Image failed to load")

            landmarks = [
                (row['ofd_1_x'], row['ofd_1_y']),
                (row['ofd_2_x'], row['ofd_2_y']),
                (row['bpd_1_x'], row['bpd_1_y']),
                (row['bpd_2_x'], row['bpd_2_y'])
            ]

            if self.augment:
                augmented = train_augs(image=img, keypoints=landmarks)
                img = augmented['image']
                landmarks = augmented['keypoints']

            img, landmarks = resize_with_landmarks(img, landmarks)
            img = np.ascontiguousarray(img).astype(np.float32) / 255.0

            img = np.stack([img, img, img], axis=0)


            landmarks = list(landmarks)

            if len(landmarks) > 4:
                landmarks = landmarks[:4]

            while len(landmarks) < 4:
                landmarks.append((None, None))

            heatmaps = []
            for (x, y) in landmarks:
                if (
                    x is None or y is None or
                    np.isnan(x) or np.isnan(y)
                ):
                    hm = np.zeros((HEATMAP_SIZE, HEATMAP_SIZE), dtype=np.float32)
                else:
                    hm = generate_heatmap(
                        HEATMAP_SIZE,
                        HEATMAP_SIZE,
                        int(x),
                        int(y),
                        GAUSSIAN_RADIUS
                    )
                heatmaps.append(hm)

            heatmaps = np.stack(heatmaps, axis=0).astype(np.float32)
            heatmaps = np.ascontiguousarray(heatmaps)

            return (
                torch.from_numpy(img.copy()).float(),
                torch.from_numpy(heatmaps.copy()).float()
            )

        except Exception as e:
            print("BAD SAMPLE:", idx, self.df.iloc[idx]['image_name'], e)

            dummy_img = np.zeros((3, 256, 256), dtype=np.float32)
            dummy_hm  = np.zeros((4, 256, 256), dtype=np.float32)

            return (
                torch.from_numpy(dummy_img).float(),
                torch.from_numpy(dummy_hm).float()
            )
```

```python
import torch
import torch.nn as nn

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch)
        )
        self.skip = nn.Conv2d(in_ch, out_ch, 1, bias=False)

    def forward(self, x):
        return torch.relu(self.conv(x) + self.skip(x))
```

```python
class DownSample(nn.Module):
    def __init__(self, ch):
        super().__init__()
        self.pool = nn.MaxPool2d(2)
        self.conv = ConvBlock(ch, ch)

    def forward(self, x):
        return self.conv(self.pool(x))


class UpSample(nn.Module):
    def __init__(self, ch):
        super().__init__()
        self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.conv = ConvBlock(ch, ch)

    def forward(self, x):
        return self.conv(self.up(x))
```

```python
class Hourglass(nn.Module):
    def __init__(self, depth, channels):
        super().__init__()
        self.depth = depth
        self.channels = channels

        self.res = ConvBlock(channels, channels)

        if depth > 1:
            self.down = DownSample(channels)
            self.inner = Hourglass(depth-1, channels)
            self.up = UpSample(channels)
        else:
            self.inner_res = ConvBlock(channels, channels)

    def forward(self, x):
        up1 = self.res(x)

        if self.depth > 1:
            low = self.down(x)
            low = self.inner(low)
            low = self.up(low)
        else:
            low = self.inner_res(x)

        return up1 + low
```

```python
class HeatmapHead(nn.Module):
    def __init__(self, ch, num_landmarks):
        super().__init__()
        self.head = nn.Sequential(
            nn.Conv2d(ch, ch, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(ch, num_landmarks, 1)
        )

    def forward(self, x):
        return self.head(x)
```

```python
class HourglassNet(nn.Module):
    def __init__(self, num_landmarks=4, channels=64, depth=4):
        super().__init__()

        self.pre = nn.Sequential(
            ConvBlock(3, channels),
            ConvBlock(channels, channels)
        )

        self.hourglass = Hourglass(depth, channels)
        self.head = HeatmapHead(channels, num_landmarks)

    def forward(self, x):
        x = self.pre(x)
        x = self.hourglass(x)
        x = self.head(x)
        return x   # (B,4,256,256)
```

```python
def train_one_epoch(model, loader, optimizer):
    model.train()
    total_loss = 0

    for imgs, heatmaps in loader:
        imgs = imgs.cuda()
        heatmaps = heatmaps.cuda()

        optimizer.zero_grad()

        preds = model(imgs)

        loss = heatmap_loss(preds, heatmaps)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
@torch.no_grad()
def validate(model, loader):
    model.eval()
    total_loss = 0

    for imgs, heatmaps in loader:
        imgs = imgs.cuda()
        heatmaps = heatmaps.cuda()

        preds = model(imgs)
        loss = heatmap_loss(preds, heatmaps)   

        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
class HybridHourglassNet(nn.Module):
    def __init__(self, num_landmarks=4, channels=64, depth=4):
        super().__init__()
        self.num_landmarks = num_landmarks

        self.pre = nn.Sequential(
            ConvBlock(3, channels),
            ConvBlock(channels, channels)
        )

        self.hourglass = Hourglass(depth, channels)

        self.hm_head = HeatmapHead(channels, num_landmarks)

        self.reg_head = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),   
            nn.Flatten(),
            nn.Linear(channels, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_landmarks * 2)  
        )

    def forward(self, x):
        feat = self.pre(x)
        feat = self.hourglass(feat)

        heatmaps = self.hm_head(feat)      
        coords = self.reg_head(feat)       
        coords = coords.view(-1, self.num_landmarks, 2)

        return heatmaps, coords
```

```python
def get_regression_targets(landmarks):

    return torch.tensor(landmarks, dtype=torch.float32)
```

```python
import os
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset

class FetalLandmarkDataset(Dataset):
    def __init__(self, df, img_dir, augment=False):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.augment = augment

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        try:
            row = self.df.iloc[idx]
            img_path = os.path.join(self.img_dir, row['image_name'])

            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                raise ValueError("Image failed to load")

            landmarks = [
                (row['ofd_1_x'], row['ofd_1_y']),
                (row['ofd_2_x'], row['ofd_2_y']),
                (row['bpd_1_x'], row['bpd_1_y']),
                (row['bpd_2_x'], row['bpd_2_y'])
            ]

            if self.augment:
                aug = train_augs(image=img, keypoints=landmarks)
                img = aug["image"]
                landmarks = aug["keypoints"]

            img, landmarks = resize_with_landmarks(img, landmarks)

            landmarks = list(landmarks)

            if len(landmarks) > 4:
                landmarks = landmarks[:4]

            while len(landmarks) < 4:
                landmarks.append((None, None))

            reg_targets = []
            for (x, y) in landmarks:
                if x is None or y is None or np.isnan(x) or np.isnan(y):
                    reg_targets.append([0.0, 0.0])
                else:
                    reg_targets.append([float(x), float(y)])

            reg_targets = np.array(reg_targets, dtype=np.float32) / HEATMAP_SIZE   # (4,2)

            img = np.ascontiguousarray(img).astype(np.float32) / 255.0
            img = np.stack([img, img, img], axis=0)   # (3,H,W)

            heatmaps = []
            for (x, y) in landmarks:
                if x is None or y is None or np.isnan(x) or np.isnan(y):
                    hm = np.zeros((HEATMAP_SIZE, HEATMAP_SIZE), np.float32)
                else:
                    hm = generate_heatmap(
                        HEATMAP_SIZE,
                        HEATMAP_SIZE,
                        int(x),
                        int(y),
                        GAUSSIAN_RADIUS
                    )
                heatmaps.append(hm)

            heatmaps = np.stack(heatmaps, axis=0).astype(np.float32)

            return (
                torch.from_numpy(img).float(),
                torch.from_numpy(heatmaps).float(),
                torch.from_numpy(reg_targets).float()
            )

        except Exception as e:
            print("BAD SAMPLE:", idx, self.df.iloc[idx]['image_name'], e)

            return (
                torch.zeros(3,256,256),
                torch.zeros(4,256,256),
                torch.zeros(4,2)
            )
```

```python
def hybrid_loss(pred_hm, gt_hm, pred_xy, gt_xy,
                w_heatmap=1.0, w_reg=0.5):

    loss_hm = F.mse_loss(pred_hm, gt_hm)

    loss_reg = F.mse_loss(pred_xy, gt_xy)

    return w_heatmap * loss_hm + w_reg * loss_reg, loss_hm.item(), loss_reg.item()
```

```python
model_hybrid = HybridHourglassNet(num_landmarks=4).cuda()
optimizer = torch.optim.Adam(model_hybrid.parameters(), lr=1e-3)

EPOCHS = 80

for epoch in range(1, EPOCHS+1):
    model_hybrid.train()
    total_loss = total_hm = total_reg = 0

    for imgs, heatmaps, gt_xy in train_loader_hybrid:
        imgs = imgs.cuda()
        heatmaps = heatmaps.cuda()
        gt_xy = gt_xy.cuda()

        pred_hm, pred_xy = model_hybrid(imgs)

        loss, hm_l, reg_l = hybrid_loss(pred_hm, heatmaps, pred_xy, gt_xy)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        total_hm += hm_l
        total_reg += reg_l
    print(f"Epoch {epoch:02d} | "
          f"Loss {total_loss/len(train_loader):.4f} | "
          f"HM {total_hm/len(train_loader):.4f} | "
          f"Reg {total_reg/len(train_loader):.4f}") 
```

```python
@torch.no_grad()
def evaluate_hybrid(n=50):
    model_hybrid.eval()
    errors_heatmap = []
    errors_reg = []

    for i, (imgs, hms, gt_xy_norm) in enumerate(val_loader_hybrid):
        if i >= n:
            break

        imgs = imgs.cuda()

        pred_hm, pred_xy_norm = model_hybrid(imgs)

        pred_xy_px = pred_xy_norm.cpu().numpy() * HEATMAP_SIZE
        gt_xy_px   = gt_xy_norm.numpy() * HEATMAP_SIZE

        pred_hm_np = pred_hm.cpu().numpy()

        for b in range(len(imgs)):
            kps_hm = []
            for k in range(4):
                y, x = np.unravel_index(pred_hm_np[b,k].argmax(),
                                        pred_hm_np[b,k].shape)
                kps_hm.append((x,y))

            for k in range(4):
                gx, gy = gt_xy_px[b,k]
                px_hx, py_hy = kps_hm[k]
                px_rg, py_rg = pred_xy_px[b,k]

                err_h = np.sqrt((gx-px_hx)**2 + (gy-py_hy)**2)
                err_r = np.sqrt((gx-px_rg)**2 + (gy-py_rg)**2)

                errors_heatmap.append(err_h)
                errors_reg.append(err_r)

    print("Heatmap  Mean:", np.mean(errors_heatmap),
          "| Median:", np.median(errors_heatmap))
    print("Regression  Mean:", np.mean(errors_reg),
          "| Median:", np.median(errors_reg))

evaluate_hybrid()
```
